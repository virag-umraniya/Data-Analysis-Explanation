## Python Arithmetic Operators

In Python, arithmetic operators are used to perform mathematical operations.  
Here are the commonly used arithmetic operators:

1. **Addition (`+`)** → Adds two numbers.  
   Example: `a + b`

2. **Subtraction (`-`)** → Subtracts the right operand from the left.  
   Example: `a - b`

3. **Multiplication (`*`)** → Multiplies two numbers.  
   Example: `a * b`

4. **Division (`/`)** → Divides left operand by right operand (result is always float).

5. **Floor Division (`//`)** → Divides and returns the integer part (quotient without decimal).  
   Example: `a // b`

6. **Modulus (`%`)** → Returns the remainder after division.  
   Example: `a % b`

7. **Exponentiation (`**`)** → Raises a number to the power of another.  
   Example: `a ** b`

# Comparison Operators in Python

Comparison operators are used to compare two values.  
They always return either **True** or **False**.

| **Operator** | **Meaning**                          | **Example**        | **Result** |
|--------------|--------------------------------------|--------------------|-------------|
| `==`         | Equal to                             | 5 == 5             | True        |
| `!=`         | Not equal to                         | 5 != 3             | True        |
| `>`          | Greater than                         | 10 > 7             | True        |
| `<`          | Less than                            | 4 < 2              | False       |
| `>=`         | Greater than or equal to             | 8 >= 8             | True        |
| `<=`         | Less than or equal to                | 6 <= 9             | True        |

**Note:**  
- These operators are mostly used in decision-making and conditional statements.  
- The output is always a **Boolean value** (`True` or `False`).


````python
# Comparison Operators in Action

# Equal to
print("5 == 5:", 5 == 5)
print("5 == 3:", 5 == 3)
print("-------------------------")

# Not equal to
print("5 != 3:", 5 != 3)
print("10 != 10:", 10 != 10)
print("-------------------------")

# Greater than
print("10 > 7:", 10 > 7)
print("3 > 9:", 3 > 9)
print("-------------------------")

# Less than
print("4 < 2:", 4 < 2)
print("2 < 5:", 2 < 5)
print("-------------------------")

# Greater than or equal to
print("8 >= 8:", 8 >= 8)
print("9 >= 10:", 9 >= 10)
print("-------------------------")

# Less than or equal to
print("6 <= 9:", 6 <= 9)
print("12 <= 5:", 12 <= 5)
````

# Logical Operators in Python

Logical operators are used to combine conditional statements.  
They also return either **True** or **False**.

| **Operator** | **Meaning**                            | **Example**              | **Result** |
|--------------|----------------------------------------|--------------------------|-------------|
| `and`        | True if **both** conditions are True   | (5 > 2) and (10 > 3)     | True        |
| `or`         | True if **at least one** condition is True | (5 > 10) or (10 > 3) | True        |
| `not`        | Reverses the result (True → False, False → True) | not(5 > 2) | False |

**Tips to Remember:**
- **and** → Both must be True
- **or** → At least one True is enough
- **not** → Opposite result


````python
# Example: And Operator

marks = 100
attendance = 90

# 1. AND Example (Both must be true)
if marks >= 40 and attendance >= 75:
    print("Yes, you passed with good attendance!")
else:
    print("No, condition not satisfied.")
````

````python
# 2. OR Example (At least one true)
assignment_done = True
extra_class_attended = False

if assignment_done or extra_class_attended:
    print("Yes, you will get bonus marks!")
else:
    print("No bonus marks.")
````

````python
# 3. NOT Example (Reverse condition)
is_shivam_late = True

print(not is_shivam_late)
````

### 2. What is a Tuple?
A **tuple** in Python is a collection of multiple items stored in a single variable.  
Tuples are **ordered, immutable (unchangeable),** and allow **duplicate values**.  
Defined using **parentheses `()`**.

#### Key Points about Tuple:
- **Ordered**: Elements have a fixed order.
- **Immutable**: Once a tuple is created, elements cannot be changed, added, or removed.
- **Allows Duplicates**: Same value can appear multiple times.
- **Faster** than lists due to immutability.
- Can be used as **dictionary keys** (since tuples are hashable).

---

#### Quick Note:
- If you need an **unchangeable collection**, use a **tuple**.
- If you need to **modify elements**, use a **list**.

````python
# Creating a Tuple
numbers = (10, 20, 30, 40, 50, 50)
print("Tuple:", numbers)
print("Type Tuple",type(numbers))
````

## 3. What is a Set?

A **Set** in Python is a collection of **unordered, unindexed, unique** elements.  
It is defined using **curly braces `{}`** or the `set()` constructor.

---

### Key Features of Set:
- **Unordered** → Elements do not have a fixed position (no indexing).
- **Unique** → Duplicate values are automatically removed.
- **Mutable** → You can add or remove elements after creation.
- **Unindexed** → Elements cannot be accessed using an index like lists or tuples.
- Supports **mathematical set operations** such as **union, intersection, difference**.
---
### Quick Note:
- Use **set** when you need a **collection of unique values**.  
- Lists and tuples allow duplicates, but **sets automatically remove them**.  


````python
# Creating a Set
numbers = {50, 20, 30, 40, 40, 10}
print(numbers)   # Duplicate 40 is removed automatically
````

````python
# Unindexed
print(numbers[0])
````

````python
# numbers[0] = 1 # This way item assignment is not possible
````

| remove(x)            | Removes `x`; gives error if not found.                | my_set.remove(2)               |
| discard(x)           | Removes `x`; **no error** if not found.               | my_set.discard(99)             |

### Set Operations

| Operation               | Symbol / Method      | Example                           |
|-------------------------|----------------------|-----------------------------------|
| Union                   | `set1 ∪ set2`        | set1.union(set2) / set1 \| set2   |
| Intersection            | `set1 ∩ set2`        | set1.intersection(set2) / set1 & set2 |
| Difference              | `set1 - set2`        | set1.difference(set2)             |
| Symmetric Difference    | `set1 △ set2`        | set1.symmetric_difference(set2)   |
| Subset Check            | -                    | set1.issubset(set2)               |

````python
# Set Operations
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
print("Union:", set1.union(set2))                  # or set1 | set2
print("Intersection:", set1.intersection(set2))    # or set1 & set2
````

## What is a Dictionary?

A **Dictionary** in Python is a collection of **key-value pairs**.  
It is defined using **curly braces `{}`** with the format:  
`key : value`

---

### Key Features of Dictionary:
- **Mutable** → Values can be changed after creation.
- **Unordered (Python < 3.7)** → No fixed order.
- **Ordered (Python ≥ 3.7)** → Insertion order is preserved.
- **Unique Keys** → Keys must be unique (duplicate keys overwrite previous value).
- **Values can be duplicate**.
- Keys must be **immutable types** (e.g., string, number, tuple).
- Values can be of **any type** (list, dict, tuple, etc.).

---

### Common Dictionary Methods

| Method                 | Description                                           | Example |
|-------------------------|-------------------------------------------------------|---------|
| get(key, default)       | Returns the value of `key`; returns `default` if not found. | d.get("name", "Unknown") |
| keys()                  | Returns all keys.                                    | d.keys() |
| values()                | Returns all values.                                  | d.values() |
| items()                 | Returns all key-value pairs as tuples.               | d.items() |
| update(dict2)           | Updates dictionary with another dictionary.          | d.update({"age": 21}) |
| pop(key)                | Removes and returns the value for the given key.     | d.pop("age") |
| popitem()               | Removes and returns the **last inserted** key-value pair. | d.popitem() |
| clear()                 | Removes all items from dictionary.                   | d.clear() |
| copy()                  | Returns a shallow copy of the dictionary.            | new_d = d.copy() |
| setdefault(key, value)  | Returns value of `key`; if not exists, inserts key with given value. | d.setdefault("city", "Delhi") |

---

### Quick Note:
- Use **dictionary** when you need to store **data with labels** (key-value mapping).
- Lists/Tuples are for sequences, **dictionaries are for mappings**.


````python
# Creating a Dictionary
student = {
    "name": "Pranav",
    "age": 10,
    "course": "BCA",
    "skills": "Python"
}
print("Dictionary:", student)
````

````python
# Create a dictionary with one key-value pair
d = {"name": "Pranav"}

# Add a new key-value pair ("age" did not exist before, so it gets added)
d["age"] = 20

# Update an existing key ("name" already exists, so its value gets changed)
d["name"] = "Harshil"

# Print the final dictionary
print(d)
````

````python
# Unique keys
d1 = {"a": 2, "a": 3, "a": 4}
print(d1)   # {"a": 2} (overwrites)
````

````python
# Value can be duplicate
d2 = {"a": 1, "b": 1, "c": 1}
print(d2)   # {"a":1, "b":1, "c":1}
````

````python
# Valid Example (Immutable keys)

# Strings as keys
dict1 = {"name": "Prashant", "course": "BCA"}
print(dict1)

# Numbers as keys
dict2 = {101: "Maths", 102: "Science"}
print(dict2)

# Tuples as keys (Tuples are immutable)
dict3 = {(1, 2): "Coordinates", (3, 4): "More Coordinates"}
print(dict3)
````

## Loops in Python

### What are Loops?
In Python, **loops** are used to repeat a block of code multiple times until a certain condition is met.  
Instead of writing the same code again and again, loops make programs shorter, cleaner, and more efficient.

There are mainly **two types of loops in Python**:

1. **while loop**  
2. **for loop**

---

### 1. While Loop
A **while loop** runs as long as the given condition is `True`.  
If the condition becomes `False`, the loop stops.

**Syntax:**
```python
while condition:
    # code to execute


````python
# Print numbers from 1 to 5 using while loop
i = 1
while i <= 5:
    print(i)
    i += 1
````

## For Loop in Python

### What is a For Loop?
A **for loop** in Python is used to iterate (loop through) a sequence of elements like:
- List
- Tuple
- String
- Range
- Dictionary

It repeats the code block once for **each element** in the sequence.

---

### Syntax:

| Feature         | While Loop                                        | For Loop                                              |
| --------------- | ------------------------------------------------- | ----------------------------------------------------- |

## Loop Control Statements in Python

In Python, **loop control statements** are used to change the normal flow of a loop.

There are 3 main control statements:
1. break
2. continue
3. pass
---

### 1. break
- Exits the loop **immediately**, even if the loop condition is still true.
- Used when we want to **stop** the loop based on a condition.

````python
for i in range(1, 10):
    if i == 5:
        break   # Exit loop when i = 5
    print(i)
````

### 2. continue Statement in Python

The **continue** statement is used to **skip the current iteration** of the loop 
and move to the next iteration.

- The loop is **not terminated**.
- Only the current cycle is skipped.

````python
for i in range(1, 6):
    if i == 3:
        continue   # Skip when i = 3
    print(i)
````

### 3. pass Statement in Python

The **pass** statement is a **null operation** — it does nothing.

- It is used as a **placeholder**.
- Keeps code syntactically correct when a statement is required, but you don’t want any action.

````python
# Example
for i in range(1, 4):
    if i == 2:
        pass   # Does nothing, just a placeholder
    print(i)
````

### Loop Control Statements in Python
| Statement    | Simple Meaning           | What it does inside the loop    | Example Use                |
| ------------ | ------------------------ | ------------------------------- | -------------------------- |
| **break**    | Stop the loop completely | Exits the loop immediately      | Stop when number becomes 5 |
| **continue** | Skip this step only      | Jumps to the next iteration     | Skip printing even numbers |
| **pass**     | Do nothing               | Keeps the loop running normally | Leave empty code for now   |

## 4. Types of Statistics
## I. Descriptive Statistics (Summarizing Data)

Descriptive statistics help you understand what the data looks like.

### A. Measures of Central Tendency

**1. Mean (Average)**
Shows the "central" value of data. Used when data is clean and balanced.

$$\text{Mean} = \frac{\sum{(\text{Values})}}{N}$$

**2. Median**
Middle value when data is sorted.
* **Useful when:** There are outliers (extreme values) or data is skewed.
* *Example:* Income of 10 people → The median is more reliable than the average.

**3. Mode**
Most frequent value.
* **Useful for:** Categorical data (finding most common category/model/color).

### B. Measures of Spread (Variability)
Tells how “spread out” or “stable” data is.

**1. Variance**
Shows how far values are from the mean.

$$\sigma^2 = \frac{\sum (x - \bar{x})^2}{N}$$

**2. Standard Deviation (SD)**
The most important measure of spread.

$$\sigma = \sqrt{\sigma^2}$$

* **Small SD:** Data is stable/consistent
* **Large SD:** Data is highly variable
* *Used in:* Manufacturing, Finance, Sales performance, Forecasting.

### C. Percentiles & Quartiles

**Percentile**
Indicates the relative position of a value.
* *Example:* If you are in the 90th percentile, you performed better than 90% of people.

**Quartiles**
Data split into 4 equal parts:
* **Q1** = 25th percentile
* **Q2** = 50th percentile (Median)
* **Q3** = 75th percentile
* *Used in:* Box plots, Detecting outliers, Customer segmentation.

### Querying with `.query()`

The `.query()` method in pandas lets you filter DataFrame rows using a string expression — it's a more readable and often more concise alternative to using boolean indexing.

This is a cleaner, SQL-like way to filter:

```python
df.query("Age > 25 and Survived == 1")
```

Here are the main **rules and tips** for using `.query()` in pandas:

---

#### **1. Column names become variables**
You can reference column names directly in the query string:

```python
df.query("age > 25 and Survived == 1")
```

---

#### **2. String values must be in quotes**
Use **single** or **double** quotes around strings in the expression:

```python
df.query("Name == 'Prashant'")
```

If you have quotes inside quotes, mix them:

```python
df.query('Name == "Prashant"')
```

---

#### **3. Use backticks for column names with spaces or special characters**
If a column name has spaces, use backticks (`` ` ``):

```python
df.query("`Name of Person` == 'Prashant'")
```

---

#### **4. You can use `@` to reference Python variables**
To pass external variables into `.query()`:

```python
age_limit = 30
df.query("Age > @age_limit")
```

---

#### **5. Logical operators**
Use these:
- `and`, `or`, `not` — instead of `&`, `|`, `~`
- `==`, `!=`, `<`, `>`, `<=`, `>=`

Bad:
```python
df.query("Age > 30 & Name == 'Prashant'")  # ❌
```

Good:
```python
df.query("Age > 30 and Name == 'Prashant'")  # ✅
```

---

#### **6. Chained comparisons**
Just like Python:

```python
df.query("25 < Age <= 40")
```

---

#### **7. Avoid using reserved keywords as column names**
If you have a column named `for`, `int`, etc., you’ll need to use backticks:

```python
df.query("`for` == 1")
```

---

#### **8. Case-sensitive**
Column names and string values are case-sensitive:

```python
df.query("Embarked == 's'")  # ❌ if actual value is 'Delhi'
```

---

#### **9. `.query()` returns a **copy**, not a view**

The result is a new DataFrame. Changes won't affect the original unless reassigned:

```python
filtered = df.query("age < 50")
```
---
